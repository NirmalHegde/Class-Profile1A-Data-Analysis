```python
pie, ax = plt.subplots(figsize=[10,6])
labels = parents.keys()
plt.pie(x=parents, autopct="%.1f%%", labels=labels, pctdistance=0.5)
plt.title("Where were SYDE 2025's parents born?", fontsize=14);
pie.savefig("outputs/parents.png")
```
