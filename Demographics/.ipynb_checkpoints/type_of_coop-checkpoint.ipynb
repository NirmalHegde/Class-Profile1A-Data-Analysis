```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
relevant_cols = ['What is your salary (in CAD/h, do not include "$")?', 'Which sector was your first co-op in?']
coop_type = class_df[relevant_cols]

coop_type.head()
```

```python
coop_type['Which sector was your first co-op in?'] = coop_type['Which sector was your first co-op in?'].str.split(';')

coop_type = (coop_type
 .set_index(['What is your salary (in CAD/h, do not include "$")?'])['Which sector was your first co-op in?']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_1', axis=1)
 .rename(columns={0:'Which sector was your first co-op in?'}))

coop_type.head(10)
```

```python
coop_type = coop_type.drop(columns=['What is your salary (in CAD/h, do not include "$")?'])
coop_type.head()
```

```python
coop_type['Number of people'] = coop_type.groupby(['Which sector was your first co-op in?'])['Which sector was your first co-op in?'].transform('count')
coop_type.head()
```

```python
coop_type = coop_type.drop_duplicates(subset=['Which sector was your first co-op in?', 'Number of people'], keep='first')
coop_type
```

```python
total_respondants = coop_type['Number of people'].sum()
coop_type['Percentage of People'] = (coop_type['Number of people'] / total_respondants) * 100
coop_type
```

```python
sns.set(font_scale=1.25)
plt.figure(figsize=(15,8))
plt.title("Types of CO-OPs", fontsize=18, y=1.02)
sns.set_theme(palette="colorblind")
ax = sns.barplot(x=coop_type["Which sector was your first co-op in?"], y=coop_type["Percentage of People"], data=coop_type)
ax.set(ylim=(0, 55))
plt.xlabel("Sectors", labelpad=15, fontsize=16)
plt.ylabel("Percentage of people (%)", labelpad=15, fontsize=16)
```
