```python
# group the ethnicities in the data frame
ethnicity_grouped = class_df.groupby("Ethnicity")["Ethnicity"].count()
print(ethnicity_grouped)
```

```python
# get the index name of each ethnicity
ethnicity_grouped.index
```

```python
ethnicity_grouped.sum()
```

```python
def adjust_ethnicites(ethnicities, num_people):
    print(ethnicities, num_people)
    # from two ethncites, split these
    if len(ethnicities) >= 2:
        for ethnicity in ethnicities:
            # add half of the people to one of the ethnicties
            if ethnicity in ethnicity_grouped.index:
                ethnicity_grouped[ethnicity] += num_people/2
            elif ethnicity != "(Mix)":
                 ethnicity_grouped[ethnicity] = num_people/2
        print(";".join(ethnicities))
        ethnicity_grouped.drop(labels=";".join(ethnicities), inplace=True)
        
# loop over indicies and adjust values accordingly
for index, num_people in ethnicity_grouped.items():
    # check if there is a ';' in the name and adjust the ethnicities accordingly
    if index.find(";"):
        ethnicities = index.split(";")
        # try to adjust the ethnicities
        adjust_ethnicites(ethnicities, num_people)

print(ethnicity_grouped)
```

# From this notebook, only upload relative percent - upload the 'percent_ethnicities' series

```python
# scale all the data to a percentage out of 100
percent_ethnicities = ethnicity_grouped.apply(lambda people: (people / ethnicity_grouped.sum() * 100).round(2))
print(percent_ethnicities.sum())
print(percent_ethnicities)
```

```python
sns.set_style('darkgrid')
ethnicities_bar = sns.barplot(x=percent_ethnicities.values, y=percent_ethnicities.index, orient='h')
ethnicities_bar.set_title("Ethnicites of SYDE Students")
ethnicities_bar.set_xlabel("Relative Percent")
ethnicities_bar.set_ylabel("Ethnicities")
```
