```python
politic_arr = politic_df.groupby("Political Alignment")["Political Alignment"].count()
politic_arr
```

```python
politic_arr_renamed = politic_arr.rename(index={"Anarchist;Libertarian": "Anarchist", "Communist;Anarchist": "Communist"})
politic_arr_renamed["Libertarian"] += 1
politic_arr_renamed["Anarchist"] += 1
politic_arr_renamed
```

```python
def group_alignment(politic):
    politic_list = politic.split(';')
    if len(politic_list) > 1:
        for i in politic_list:
            politic_arr_renamed[i] += politic_arr_renamed[politic]
        politic_arr_renamed.drop(labels=politic, inplace=True)

for politic in politic_arr_renamed.index:
    group_alignment(politic)

politic_arr_renamed
```

```python
politic_arr_renamed = politic_arr_renamed.astype('float64')
politic_arr_renamed['Communist'] += 1
politic_arr_renamed
```

```python
total_choices = 0
for politic in politic_arr_renamed:
    total_choices += politic

for i in range(len(politic_arr_renamed)):
    politic_arr_renamed[i] = politic_arr_renamed[i] / total_choices * 100

politic_arr_renamed
```

```python
plt.title("Political Alignment of SYDE Students")
sns.set_theme(palette="colorblind")
sns.barplot(x=politic_arr_renamed.values, y=politic_arr_renamed.index, orient='h')
plt.xlabel("Percentage of Choices (%)")
plt.ylabel("Political Alignment")
```
