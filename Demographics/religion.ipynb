```python
religion_df['Number of people'] = religion_df.groupby(['Religion'])['Religion'].transform('count')
religion_df = religion_df.drop_duplicates(subset=['Religion', 'Number of people'], keep='first')

religion_df
```

```python
total_respondants = religion_df['Number of people'].sum()
religion_df['Percentage of people'] = (religion_df['Number of people'] / total_respondants) * 100
religion_df = religion_df.reset_index()
religion_df
```

```python
plt.figure(figsize=(19,10))
plt.title("Religions In the SYDE 2025 Cohort", fontsize=16, y=1.02)
sns.set_theme(palette="colorblind")
ax = sns.barplot(x=religion_df['Religion'],y=religion_df['Percentage of people'], data=religion_df)
ax.set(ylim=(0, 25), )
plt.xlabel("Religions", labelpad=15, fontsize=16)
plt.ylabel("Percentage of Respondents (%)", labelpad=15, fontsize=16)
```
