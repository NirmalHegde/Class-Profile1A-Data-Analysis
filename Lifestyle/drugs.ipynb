```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
relevant_cols = ['What is your favourite programming language?','If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)']
drugs = class_df[relevant_cols]

drugs.head()
```

```python
drugs['If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)'] = drugs['If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)'].str.split(';')

drugs = (drugs
 .set_index(['What is your favourite programming language?'])['If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_1', axis=1)
 .rename(columns={0:'If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)'}))


drugs.head(10)
```

```python
drugs = drugs.drop(columns=['What is your favourite programming language?'])
drugs.head()
```

```python
drugs['Number of people'] = drugs.groupby(['If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)'])['If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)'].transform('count')
drugs.head()
```

```python
drugs = drugs.drop_duplicates(subset=['If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)', 'Number of people'], keep='first')
drugs
```

```python
drugs.loc[drugs['If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)'] == 'Cigarretes','If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)'] = 'Cigarettes'

drugs.head(10)
```

```python
total_respondants = drugs['Number of people'].sum()
drugs['Percentage of People'] = (drugs['Number of people'] / total_respondants) * 100
drugs
```

```python
plt.figure(figsize=(15,10))
plt.title("Types of Drugs People Have Used", fontsize=18, y=1.03)
sns.set_theme(palette="colorblind")
sns.set(font_scale=1.2)
ax = sns.barplot(x=drugs["Number of people"], y=drugs["If you have used recreational drugs, which ones? (If you have not used drugs, indicate as such)"], data=drugs)
ax.tick_params(labelsize=14)
plt.xlabel("Percentage of People (%)", labelpad=15, fontsize=16)
plt.ylabel("Type of drug", labelpad=15, fontsize=16)
```
