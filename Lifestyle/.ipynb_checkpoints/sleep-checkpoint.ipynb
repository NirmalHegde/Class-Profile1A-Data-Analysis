```python
sleep['Number of people'] = sleep.groupby(['How many hours of sleep did you get per night during school weeks on average (Not including naps)'])['How many hours of sleep did you get per night during school weeks on average (Not including naps)'].transform('count')

sleep.head()
```

```python
sleep = sleep.drop_duplicates(subset=['How many hours of sleep did you get per night during school weeks on average (Not including naps)', 'Number of people'], keep='first')

sleep
```

```python
sleep = sleep.dropna()
sleep
```

```python
total_respondants = sleep['Number of people'].sum()
sleep['Percentage of People'] = (sleep['Number of people'] / total_respondants) * 100
sleep.drop(['Number of people'], axis=1, inplace=True)
sleep = sleep.reset_index()
sleep
```

```python
plt.figure(figsize=(15,8))
plt.title("Average Hours of Sleep Per Night During School Weeks", fontsize=18, y=1.02)
sns.set_theme(palette="colorblind")
ax = sns.barplot(x=sleep["How many hours of sleep did you get per night during school weeks on average (Not including naps)"], y=sleep["Percentage of People"], data=sleep)
ax.tick_params(labelsize=14)
ax.set(ylim=(0, 50))
plt.xlabel("Number of hours of sleep per night", labelpad=15, fontsize=16)
plt.ylabel("Percentage of the cohort (%)", labelpad=15, fontsize=16)
```
