```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
# Comparing mental health against grades
relevant_cols = ['Rate your mental health during the term.', 
'What was your final grade for 101?', 
'What was your final grade for 101L?', 
'What was your final grade for 111? ',
'What was your final grade for 113?', 
'What was your final grade for 121?', 
'What was your final grade for 161?', 
'What was your final grade for 181?']
mhealth_v_grades = class_df[relevant_cols]

mhealth_v_grades
```

```python
def get_average(row):
   return (row['What was your final grade for 101?']*0.5 + row['What was your final grade for 101L?']*0.5 
   + row['What was your final grade for 113?']   *0.5 + row['What was your final grade for 121?'] + row['What was your final grade for 161?'] 
   + row['What was your final grade for 181?'] + row['What was your final grade for 111? '])/5.5

mhealth_v_grades['Average'] = mhealth_v_grades.apply(get_average, axis=1)
mhealth_v_grades
```

```python
# creating scatterplot to show mental health vs grades
sns.set(rc={'figure.figsize':(10,6)})
ax=sns.scatterplot(data=mhealth_v_grades, x="Rate your mental health during the term.", y="Average")
ax.set_ylabel("Average during 1A")
ax.set_xlabel("Mental health during 1A")
ax.set_title("Sleep vs Mental Health")
```

```python
# applying linear regression to scatterplot
sns.set(rc={'figure.figsize':(10,6)})
ax=sns.regplot(x="Rate your mental health during the term.", y="Average", data=mhealth_v_grades)
ax.set_ylabel("Average during 1A")
ax.set_xlabel("Mental health during 1A")
ax.set_title("Sleep vs Mental Health")
```

```python
# applying parabolic regression
sns.set(rc={'figure.figsize':(10,6)})
ax=sns.regplot(x="Rate your mental health during the term.", y="Average", data=mhealth_v_grades, order=2)
ax.set_ylabel("Average during 1A")
ax.set_xlabel("Mental health during 1A")
ax.set_title("Sleep vs Mental Health")
```

```python
# applying cubic regression
sns.set(rc={'figure.figsize':(10,6)})
ax=sns.regplot(x="Rate your mental health during the term.", y="Average", data=mhealth_v_grades, order=3)
ax.set_ylabel("Average during 1A")
ax.set_xlabel("Mental health during 1A")
ax.set_title("Sleep vs Mental Health")
```

```python
# applying quartic regression
sns.set(rc={'figure.figsize':(10,6)})
ax=sns.regplot(x="Rate your mental health during the term.", y="Average", data=mhealth_v_grades, order=4)
ax.set_ylabel("Average during 1A")
ax.set_xlabel("Mental health during 1A")
ax.set_title("Mental Health vs Average During 1A")
ax.set(ylim=(60,100))
```

```python
sns.set(rc={'figure.figsize':(15, 9)})
fig, ax = plt.subplots()
sns.boxplot(x="Rate your mental health during the term.", y="Average", data=mhealth_v_grades, ax=ax, width=0.2)
plt.title("Grades vs Mental Health", fontsize=18, y=1.03)
plt.xlabel("Mental health during 1A (Rating Out of 10)", labelpad=15, fontsize=16)
plt.ylabel("Average during 1A (%)", labelpad=15, fontsize=16)

plt.show()
```
