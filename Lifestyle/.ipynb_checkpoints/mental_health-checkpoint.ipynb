```python
mhealth = pd.DataFrame(columns=["Rating Out of 10", "Before 1A", "During 1A", "After 1A"],
    data=[["1", 0, 2, 1], 
        ["2", 0, 5, 1],
        ["3", 3, 2, 2],
        ["4", 6, 5, 5],
        ["5", 5, 11, 6],
        ["6", 19, 19, 9],
        ["7", 18, 20, 26],
        ["8", 27, 14, 20],
        ["9", 19, 6, 18],
        ["10", 6, 0, 6],
        ])

mhealth = pd.melt(mhealth, id_vars="Rating Out of 10", var_name="Period of Time", value_name="Number of People")
mhealth
```

```python
mhealth['Percentage of People'] = (mhealth['Number of People'] / 94) * 100
mhealth
```

```python
sns.set(font_scale = 1.25)
sns.set_theme(palette="colorblind")
ax=sns.catplot(x='Rating Out of 10', y='Percentage of People', hue='Period of Time', data=mhealth, kind='bar', height=8, aspect=1.6)
ax.set(ylim=(0, 30))
plt.title("Mental Health Before, During and After 1A", fontsize=18, y=1.03)
plt.xlabel("Mental health rating out of 10", labelpad=15, fontsize=16)
plt.ylabel("Percentage of people (%)", labelpad=15, fontsize=16)
```
