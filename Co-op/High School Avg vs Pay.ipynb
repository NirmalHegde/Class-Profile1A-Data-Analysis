```python
cols = ['What was your top 6 highschool average for engineering? Please enter just the number with NO % sign (Calc, Adv Functions, English, Physics, Chemistry, 6th course in your top 6) ',
       'What is your salary (in CAD/h, do not include "$")?']

df = class_df[cols].rename(columns={'What was your top 6 highschool average for engineering? Please enter just the number with NO % sign (Calc, Adv Functions, English, Physics, Chemistry, 6th course in your top 6) ': "avg",
                                   'What is your salary (in CAD/h, do not include "$")?': 'salary'})

df.head(20)
```

```python
# remove all entries with no avg
df = df[df['avg'].notnull()]
# remove all entries with no salary
df = df[df['salary'].notnull()]

df.head(20)
```

```python
def clean_avg_and_salary(row):
    if row['avg'] % 1 >= 0.5:
        row['avg'] += 1
    row['avg'] = int(row['avg'])
    
    if row['salary'] > 30 or row['salary'] < 13:
        row['salary'] = None
    
    return row

df = df.apply(clean_avg_and_salary, axis=1).dropna()
df.head(20)
        
```

```python
df['avg'].corr(df['salary'])
```

```python
len(df)
```

```python
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize': (7.5, 12)})
# plot salary and average in a scatter plot
ax = sns.lmplot(data=df, x="avg", y="salary")
ax.axes.flatten()[0].set_title("Admission Average vs Coop Salary", fontsize=20)
ax.set_axis_labels(x_var="Admission Average", y_var="Coop Salary (CAD)", fontsize=14)
ax.set(xlim=(84.5, 100))
```

```python
ax.savefig('graphs/coop_pay_v_admission_avg.png')
```

```python
# get the data for 1st sem averages and make a similar graph
avg_cols = [col for col in class_df.columns if col.startswith("What was your final grade for")]
salary_and_avg = avg_cols + ['What is your salary (in CAD/h, do not include "$")?']
df2 = class_df[salary_and_avg].rename(columns={'What was your final grade for 101?': '101',
                                              'What was your final grade for 101L?': '101L',
                                              'What was your final grade for 111? ': '111',
                                              'What was your final grade for 113?': '113',
                                              'What was your final grade for 121?': '121',
                                              'What was your final grade for 161?': '161',
                                              'What was your final grade for 181?': '181',
                                              'What is your salary (in CAD/h, do not include "$")?': 'salary'})
df2.head()
```

```python
def clean_df(row):
    if row['salary'] > 40 or row['101'] == np.nan or row['salary'] < 10:
        return None
    row['avg'] = ((row['101'] + row['101L'] + row['113']) * 0.25 
                  + (row['111'] + row['121'] + row['161'] + row['181']) * 0.5) / 275 * 100
    # print(type(row))
    return row

df2 = df2.apply(clean_df, axis=1).dropna()
df2.drop(columns=['101', '101L', '111', '113', '121', '161', '181'], inplace=True)
df2.head(50)
```

```python
def group_1A_avg(row):
    # group the grades in the following ranges: 95-100, 90-95, 85-90, 80-85, >80
    if row['avg'] >= 95:
        row['avg'] = 95
    elif row['avg'] >= 90:
        row['avg'] = 90 
    elif row['avg'] >= 85:
        row['avg'] = 85
    elif row['avg'] >= 80:
        row['avg'] = 80
    else:
        row['avg'] = 75
    return row

df2 = df2.apply(group_1A_avg, axis=1)
df2
```

```python
term_avg_v_coop = sns.boxplot(data=df2, x='avg', y='salary')
term_avg_v_coop.set_xticklabels(['70-79%', '80-84%', '85-89%', '90-94%', '95-100%'])
term_avg_v_coop.set_xlabel('1A Average')
term_avg_v_coop.set_ylabel('Salary (CAD)')
term_avg_v_coop.set_title('1A Average vs Coop Salary', fontsize=20)
```
