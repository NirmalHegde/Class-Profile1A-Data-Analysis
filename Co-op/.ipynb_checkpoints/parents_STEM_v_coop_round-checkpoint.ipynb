```python
relevant_cols = ['Did at least one of your parents pursue a degree/career in STEM?', 'Which round did you get your coop in?']
parents_round = class_df[relevant_cols]
parents_round.head(30)
```

```python
parents_round = parents_round[parents_round['Which round did you get your coop in?'].notna()]
parents_round.head(30)
```

```python
# filtering for people who have parents in STEM
stem_parents = parents_round[parents_round['Did at least one of your parents pursue a degree/career in STEM?'] == 'Yes']
stem_parents
```

```python
stem_parents['Number of people'] = stem_parents.groupby(['Which round did you get your coop in?'])['Did at least one of your parents pursue a degree/career in STEM?'].transform('count')
stem_parents
```

```python
stem_parents = stem_parents.drop_duplicates(subset=['Did at least one of your parents pursue a degree/career in STEM?', 'Which round did you get your coop in?', 'Number of people'], keep='first')
stem_parents
```

```python
# filtering for people who dont have parents in STEM
nonstem_parents = parents_round[parents_round['Did at least one of your parents pursue a degree/career in STEM?'] == 'No']
nonstem_parents
```

```python
nonstem_parents['Number of people'] = nonstem_parents.groupby(['Which round did you get your coop in?'])['Did at least one of your parents pursue a degree/career in STEM?'].transform('count')
nonstem_parents
```

```python
nonstem_parents = nonstem_parents.drop_duplicates(subset=['Did at least one of your parents pursue a degree/career in STEM?', 'Which round did you get your coop in?', 'Number of people'], keep='first')
nonstem_parents
```

```python
stemParents = pd.DataFrame(columns=["Round", "With STEM Parents", "Without STEM Parents"],
    data=[["First Round", 13, 4], 
        ["Second Round", 18, 15],
        ["Continuous Round", 24, 15],
        ["Direct Offer Round", 6, 4]])

stemParents = pd.melt(stemParents, id_vars="Round", var_name="Parents", value_name="Number of People")
stemParents
```

```python
total_respondants = stemParents['Number of People'].sum()
total_respondants
stemParents['Percentage of People'] = (stemParents['Number of People'] / total_respondants) * 100
stemParents
```

```python
sns.set(font_scale=1.2)
sns.set_theme(palette="colorblind")
ax=sns.catplot(x='Round', y='Percentage of People', hue='Parents', data=stemParents, kind='bar', height=8, aspect=1.5)
ax.set(ylim=(0, 30))
plt.title("Parents In STEM VS CO-OP Round", fontsize=18, y=1.03)
plt.xlabel("Round", labelpad=15, fontsize=16)
plt.ylabel("Percentage of people (%)", labelpad=15, fontsize=16)
```
