```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
class_df.columns
```

```python
relevant_cols = ['Which round did you get your coop in?', 'Which sector was your first co-op in?']
round_coop = class_df[relevant_cols]
round_coop.head(30)
```

```python
round_coop = round_coop[round_coop['Which round did you get your coop in?'].notna()]
round_coop.head(30)
```

```python
round_coop['Which sector was your first co-op in?'] = round_coop['Which sector was your first co-op in?'].str.split(';')

round_coop = (round_coop
 .set_index(['Which round did you get your coop in?'])['Which sector was your first co-op in?']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_1', axis=1)
 .rename(columns={0:'Which sector was your first co-op in?'}))

round_coop.head(50)
```

```python
round_coop[round_coop['Which sector was your first co-op in?'] != 'Unhired']
round_hired_only = round_coop[round_coop['Which sector was your first co-op in?'] != 'Unhired']
round_hired_only.head(25)
```

```python
# checking for sectors in the direct offer round
direct_offer = round_coop[round_coop['Which round did you get your coop in?'] == 'Direct Offer']
direct_offer
```

```python
direct_offer['Number of people'] = direct_offer.groupby(['Which sector was your first co-op in?'])['Which round did you get your coop in?'].transform('count')
direct_offer
```

```python
direct_offer = direct_offer.drop_duplicates(subset=['Which round did you get your coop in?', 'Which sector was your first co-op in?', 'Number of people'], keep='first')
direct_offer
```

```python
# displaying specific sectors that people got a coop in during the direct offer round
plt.figure(figsize=(16,8))
ax1 = plt.subplot(121, aspect='equal')
plt.title("Direct Offer Round Co-op Sectors", fontsize=18)
direct_offer.plot(kind='pie', y ='Number of people', ax=ax1, autopct='%1.1f%%', 
startangle=90, shadow=False, ylabel = '', labels=direct_offer['Which sector was your first co-op in?'], legend = False, fontsize=14)
```
