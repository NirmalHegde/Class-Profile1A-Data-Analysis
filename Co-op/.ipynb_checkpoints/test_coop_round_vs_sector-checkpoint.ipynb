Round 1 Co-op

```python
round1_coop = round1_coop[round1_coop['Which round did you get your coop in?'].notna()]
round1_coop.head()
```

```python
round1_coop['Which sector was your first co-op in?'] = round1_coop['Which sector was your first co-op in?'].str.split(';')

round1_coop = (round1_coop
 .set_index(['Which round did you get your coop in?'])['Which sector was your first co-op in?']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_1', axis=1)
 .rename(columns={0:'Which sector was your first co-op in?'}))

round1_coop.head()
```

```python
round1_coop[round1_coop['Which sector was your first co-op in?'] != 'Unhired']
round1_coop = round1_coop[round1_coop['Which sector was your first co-op in?'] != 'Unhired']
round1_coop.head()
```

```python
# checking for sectors in the first round
round1_coop = round1_coop[round1_coop['Which round did you get your coop in?'] == '1st Round']
round1_coop.head()
```

```python
round1_coop['Number of people'] = round1_coop.groupby(['Which sector was your first co-op in?'])['Which round did you get your coop in?'].transform('count')

round1_coop = round1_coop.drop_duplicates(subset=['Which round did you get your coop in?', 'Which sector was your first co-op in?', 'Number of people'], keep='first')

round1_coop
```
