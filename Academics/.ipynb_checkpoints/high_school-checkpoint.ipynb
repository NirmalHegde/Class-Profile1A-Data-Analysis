```python
# drop the empty grades
filtered_grades = high_school_grades.dropna().sort_values(ascending=False)
print(filtered_grades[-10:])
print(filtered_grades[:10])
```

```python
len(filtered_grades)
```

```python
sns.set_style('darkgrid')
hs_avg_kde = sns.kdeplot(data=filtered_grades, shade=True, label="High School Average", clip=(0, 100))
hs_avg_kde.set(xlim=(80, 100))
hs_avg_kde.set_xlabel('High school average')
hs_avg_kde.set_ylabel('Frequency')
hs_avg_kde.set_title('High school admission avg KDE')
```

```python
hs_avg_kde.figure.savefig('graphs/high_school_kde.png')
```

```python
sns.set_style('darkgrid')
sns.set(rc={'figure.figsize': (5, 8)})
hs_avg_graph = sns.histplot(filtered_grades, kde=True, bins=20)
hs_avg_graph.set(xlim=(83, 100))
hs_avg_graph.set_xlabel('Admission Average')
hs_avg_graph.set_title('Admission Average Histogram', fontsize=20)
```

```python
hs_avg_graph.figure.savefig('graphs/high_school_histogram.png')
```

```python
mean_grade = filtered_grades.mean()
median_grade = filtered_grades.median()

print(f"Mean Entrance Average: {mean_grade}")
print(f"Median Entrance Average: {median_grade}")
```
