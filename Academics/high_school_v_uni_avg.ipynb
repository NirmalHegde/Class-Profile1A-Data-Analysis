```python
def bin_grade_deciles(uni_1A_avg):
    if uni_1A_avg >= 90:
        uni_1A_avg = 90
    elif uni_1A_avg >= 80:
        uni_1A_avg = 80
    elif uni_1A_avg >= 70:
        # print(uni_1A_avg)
        uni_1A_avg = 70
    else:
        # print(uni_1A_avg)
        uni_1A_avg = 60
    return uni_1A_avg

uni_deciles = student_course_means.apply(bin_grade_deciles)
uni_deciles
```

```python
# round all high school admission averages to an integer
def round_grades(val):
    # check if there is a decimal percent and it and the decimal is greater than 0.5, if so round up
    decimal = val%1
    if decimal != 0 and decimal >= 0.5:
        val += 1
    return int(val)

   
rounded_grades = filtered_grades.apply(round_grades)

print(rounded_grades)
```

```python
def group_grades(admission_average):
    # group the grades in the following ranges: 85-89, 90-93, 94-96, 97-99
    if admission_average <= 89:
        admission_average = 89
    elif admission_average <= 93:
        admission_average = 93
    elif admission_average <= 96:
        admission_average = 96
    else:
        admission_average = 99
    
    return admission_average

# apply the grouping function to all grades
grouped_grades = rounded_grades.apply(group_grades)
grouped_grades.head(20)
```

```python
# create a uni vs high school graph for the newly grouped grades
labels = ["85-89", "90-93", "94-96", "97-99"]
sns.set(rc={'figure.figsize': (5, 8)})
ax = sns.boxplot(x=grouped_grades.values, y=student_course_means.values)
ax.set_title("High School Admission Average vs Uni Marks", fontsize=20)
ax.set_xlabel("Admission Average")
ax.set_ylabel("1A Average")
ax.set_xticklabels(labels)
```
