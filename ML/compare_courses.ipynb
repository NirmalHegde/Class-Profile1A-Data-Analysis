## Compare courses using PCA and TSNE
- 4 columns will be averages for each course: Average, usefulness, difficults, and % of lectures watched
- PCA will be applied with the data normalized, un-normalized and both will be plotted
- TSNE will be applied with the data normalized, un-normalized and both will be plotted

```python
import pandas as pd  
import numpy as np 
import seaborn as sns
%matplotlib inline
```

```python
# this cell has not been run to preserve sensitive info from the survey
class_df = pd.read_csv('../SYDE 1A Class Profile Questions.csv')
```

```python
course_list = ['101', '101L', '111', '113', '121', '161', '181']
all_course_cols = []
for course in course_list:
    current_course_cols = [col for col in class_df.columns if course in col and col not in all_course_cols]
    all_course_cols += current_course_cols
all_course_cols
```

```python
# confirm that we have 28 columns; 7 course x 4 columns
len(all_course_cols)
```

```python
course_df = class_df[all_course_cols]
course_df.head()
```

## Clean the data
- The data is not clean; apply function to turn all categorical data in numerical data in order to apply PCA and TSNE

```python
def handle_course_cols(row):
    """
    Turn each of the categorical columns that have to do with courses into integers either through mappings or string manipulation
    """
    # deal with the empty 181 by placing it as a value of 2 (same as neutral)
    usefulness_mappings = {'Useless': 0, 'Somewhat Useless': 1,'Neutral': 2, 'Somewhat Useful': 3,'Very Useful': 4, np.NaN: 2}
    difficulty_mappings = {'Easy': 4, 'Somewhat Easy': 3, 'Neutral': 2, 'Somewhat Difficult': 1, 'Very Difficult': 0} 
    class_list = ['[101]', '[101L]', '[111]', '[113]', '[121]', '[161]', '[181]']   

    for syde_class in class_list:
        row[f'How useful did you find each of the 1A courses to be? {syde_class}'] = usefulness_mappings[row.get(f'How useful did you find each of the 1A courses to be? {syde_class}')]
        row[f'How difficult did you find each 1A course to be? {syde_class}'] = difficulty_mappings[row.get(f'How difficult did you find each 1A course to be? {syde_class}')]

        if type(row[f'Roughly what percentage of lectures did you watch for each course?? {syde_class}']) != float:
             row[f'Roughly what percentage of lectures did you watch for each course?? {syde_class}'] = int(row.get(f'Roughly what percentage of lectures did you watch for each course?? {syde_class}')[:-1])
        else:
            row[f'Roughly what percentage of lectures did you watch for each course?? {syde_class}'] = 50
    
    return row
```

```python
course_df = course_df.apply(handle_course_cols, axis=1)
course_df.head()
```

```python
# confirm that all the data is numerical by looking at the info
course_df.info()
```

## Group by course
- In order to apply PCA/TSNE for each course, data must be aggregated by course (median)
- Also aggreate course by mean to see if there is a difference in results

```python
median_course_series = course_df.apply(lambda col: col.median())
median_course_series
```

```python
mean_course_series = course_df.apply(lambda col: col.mean())
mean_course_series
```

## Manipulate the data so that it is in the form of a df
- One column of the df should be the course and the otehr columns should be the 4 course catgeories
- After this, data will be ready for TSNE and PCA use

```python
col_list = [
    'What was your final grade',
    'How useful did you find',
    'How difficult did you find',
    'Roughly what percentage of lectures did you watch'
]

median_course_df = pd.DataFrame(index=course_list, columns=col_list)
mean_course_df = pd.DataFrame(index=course_list, columns=col_list)

mean_course_df
```

```python
def create_df_from_series(row, df):
    for course in course_list:
        for col in col_list:
            if course in row.name and col in row.name and pd.isnull(df[col][course]):
                df[col][course] = row[0]

pd.DataFrame(median_course_series).apply(create_df_from_series, axis=1, args=(median_course_df ,))
median_course_df
```

```python
pd.DataFrame(mean_course_series).apply(create_df_from_series, axis=1, args=(mean_course_df ,))
mean_course_df
```

## Apply PCA and TSNE
- First, just apply it on un-normalized data

```python
from sklearn.decomposition import PCA

def create_pca(df: pd.DataFrame) -> PCA:
    pca = PCA(n_components=2)
    pca_model = pd.DataFrame(pca.fit_transform(df))
    pca_model = pca_model.join(pd.DataFrame(course_list, columns=['course']))
    return pca_model
```

```python
def plot_courses(df: pd.DataFrame, title: str) -> sns.scatterplot:
    """ Graph the courses after PCA/TSNE was applied

    Args:
        df: The datafame containing the dimensions and the corresponding course labels
        title: The title for the graph
    
    Returns:
        A scatterplot of the data
    """
    try:
        ax = sns.scatterplot(x=df[0], y=df[1], hue=df['course'])
        ax.set_xlabel('Dim 1')
        ax.set_ylabel('Dim 2')
        ax.set_title(title)

        return ax
    except Exception as e:
        print(f'Error {e}')
```

```python
# use PCA on unnormalized averages of each course
un_norm_mean_pca = create_pca(mean_course_df)

un_norm_mean_pca
```

```python
ax_1 = plot_courses(un_norm_mean_pca, 'PCA for averages of un-normalized courses')
```

```python
# use PCA on unnormalized medians of each course
un_norm_median_pca = create_pca(median_course_df)

un_norm_median_pca
```

```python
ax_2 = plot_courses(un_norm_median_pca, 'PCA for medians of un-normalized courses')
```

```python
from sklearn.manifold import TSNE

def create_tsne(df: pd.DataFrame) -> TSNE:
    tsne = TSNE(n_components=2, n_jobs=-1)
    tsne_model = pd.DataFrame(tsne.fit_transform(df))
    tsne_model = tsne_model.join(pd.DataFrame(course_list, columns=['course']))
    return tsne_model
```

```python
# use TSNE on unnormalized averages of each course
un_norm_mean_tsne = create_tsne(mean_course_df)

un_norm_mean_tsne
```

```python
ax_3 = plot_courses(un_norm_mean_tsne, 'TSNE for averages of un-normalized courses')
```

```python
# use tsne on unnormalized medians of each course
un_norm_median_tsne = create_tsne(median_course_df)

un_norm_median_tsne
```

```python
ax_4 = plot_courses(un_norm_median_tsne, 'TSNE for medians of un-normalized courses')
```

## Normalize the data
- PCA works with difference in spread of the data so it is importance to normalize the results
- see what comes out differently for tsne as well when normalizing results

```python
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.pipeline import Pipeline

pca_pipeline = Pipeline(steps=[
    ('scale_data', StandardScaler()),
    ('labelled_pca', FunctionTransformer(create_pca))
])
```

```python
mean_pca_scaled = pca_pipeline.fit_transform(mean_course_df)
mean_pca_scaled
```

```python
ax_5 = plot_courses(mean_pca_scaled, 'PCA averages of scaled courses')
```

```python
median_pca_scaled = pca_pipeline.fit_transform(median_course_df)
median_pca_scaled
```

```python
ax_6 = plot_courses(median_pca_scaled, 'PCA median for scaled courses')
```

```python
from sklearn.base import TransformerMixin, BaseEstimator

class Debug(BaseEstimator, TransformerMixin):

    def transform(self, X):
        # store the attribute X_ to see scaled data in this step
        self.X_ = X
        return X
        
    def fit(self, X, y=None, **fit_params):
        return self
```

```python
tsne_pipeline = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('debug', Debug()),
    ('labelled_tsne', FunctionTransformer(create_tsne))
])
tsne_pipeline
```

```python
mean_tsne_scaled = tsne_pipeline.fit_transform(mean_course_df)
# see how the scaled data looks in the tsne pipeline
tsne_pipeline['debug'].X_
```

```python
mean_tsne_scaled
```

```python
ax_7 = plot_courses(mean_tsne_scaled, 'TSNE averages of scaled courses')
```

```python
median_tsne_scaled = tsne_pipeline.fit_transform(median_course_df)
median_tsne_scaled
```

```python
ax_8 = plot_courses(median_tsne_scaled, 'TSNE medians of scaled courses')
```

## Interpret Results
- TSNE both scaled and unscaled does not reveal any similarities between the courses; in fact they are all quite sparse
- After scaling PCA, it seems like the variation in data is similar for 2 groups of courses (below)
- PCA group 1: 101 and 101L are close to each other in both mean and median calculations per course. This means that the ends (the axis where most of the data lies; the variation) of the data was spread similarly among both courses. These were both meant to be communication courses so similarity in feature distribution makes sense - both courses required communication skills. 
- PCA group 2: 113 and 181 are also close to each other in the mean pca calculations, and are relatively close in median calculations. These results also make sense because they were both math/physics courses.
